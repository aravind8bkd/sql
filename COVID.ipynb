```python
import pandas as pd
import sqlite3
import os
```

```python
# Define the path for your SQLite database file
db_path = '/kaggle/working/covid.db'

# Connect to the SQLite database (it will create the file if it doesn't exist)
conn = sqlite3.connect(db_path)
```

```python
# Path to the folder containing your files
folder_path = '/kaggle/input/covid-xl/'

# Loop through all CSV files in the folder
for file in os.listdir(folder_path):
    if file.endswith('.xlsx'):
        # Read the CSV file into a DataFrame
        df = pd.read_excel(os.path.join(folder_path, file))
        
        # Use the filename (without extension) as the table name
        table_name = os.path.splitext(file)[0]
        
        # Write the DataFrame to the SQLite database
        df.to_sql(table_name, conn, if_exists='replace', index=False)

# Close the connection to the database
conn.close()
```

```python
# Get the list of tables
tables = %sql SELECT name FROM sqlite_master WHERE type='table'; 
```

```python
# Iterate over tables and print schema
for table in tables:
    table_name = table[0]
    print(f"\nSchema for table: {table_name}")
    
    # Get schema information for the current table
    query = f"PRAGMA table_info({table_name});"
    schema = %sql {query}
    
    # Display schema with headers
    display(schema)
```
